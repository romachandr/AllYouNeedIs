```python
import numpy as np
from matplotlib import pyplot as plt
import warnings

%matplotlib inline
```

```python
warnings.filterwarnings('ignore')
```

```python
    %%latex
    
    ALL YOU NEED IS...
    $$ $$
    
    $$y = \frac{1}{x}$$
    $$ $$
    $$x^2 + y^2 = 9$$
    $$ $$
    $$y = 3|x|$$
    $$ $$
    $$x = -3|sin(x)|$$
```

```python
def l_func(x):
    return 1/x

def o_func_1(x):
    return 1.5*np.sqrt(9-x**2)

def o_func_2(x):
    return -1.5*np.sqrt(9-x**2)

def v_func(x):
    return 3.5*np.abs(x)

def e_func(x):
    return -3*np.abs(np.sin(x))
```

```python
# L
x_first = 0e-5
x_last = 1.75
x = np.linspace(x_first, x_last, num=100)

y = list(map(l_func, x))

fig = plt.figure(figsize=(20, 7))
ax = fig.add_subplot(1, 4, 1)
ax.plot(x, y, 'r', linewidth=4)
ax.grid(True)
plt.xlim(0, 2)

# O
x_first = -3
x_last = 3
x = np.linspace(x_first, x_last, num=100)

y = list(map(o_func_1, x))
y_minus = list(map(o_func_2, x))

ax = fig.add_subplot(1, 4, 2)
ax.plot(x, y, 'r', linewidth=4)
ax.plot(x, y_minus, 'r', linewidth=4)
ax.grid(True)
plt.xlim(-3.5, 3.5)
plt.ylim(-5.0, 5.0)

# V
x_first = -1
x_last = 1
x = np.linspace(x_first, x_last, num=100)

y = list(map(v_func, x))

ax = fig.add_subplot(1, 4, 3)
ax.plot(x, y, 'r', linewidth=4)
ax.grid(True)
plt.xlim(-1.25, 1.25)
plt.ylim(-0.25, 3.75)

# E
x_first = -np.pi
x_last = np.pi
x = np.linspace(x_first, x_last, num=100)

y = list(map(e_func, x))

ax = fig.add_subplot(1, 4, 4)
ax.plot(y, x, 'r', linewidth=4)
ax.grid(True)

# ---
plt.suptitle('ALL YOU NEED IS...', fontsize=30)
plt.show()
```
